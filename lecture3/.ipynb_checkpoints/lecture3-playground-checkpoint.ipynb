```r
diamonds  %>% head
```

```r
diamonds %>% ggplot(aes(x = carat, y = price)) + geom_point()
```

```r
# adding color
diamonds %>% ggplot(aes(x = carat, y = price)) + geom_point(aes(color = color), alpha = 0.2)
```

```r
# adding another layer
diamonds %>% ggplot(aes(x = carat, y = price)) + geom_point(aes(color = color), alpha = 0.2) + geom_smooth()
```

```r
# this one is different
ggplot(diamonds,
aes(x = carat, y = price, col = color),
alpha = 0.2) +
geom_point() +
geom_smooth()
```

**Arguments to a ggplot call:**
* The first argument to a ggplot is the data frame (or tibble)
* You can fix the aesthetics with aes() OR you can add
aesthetics in layers
* **Any aspect of a plot that you want to vary based on a variable
needs to be wrapped in a aes() call**

```r
# more layers
diamonds  %>% ggplot(aes(x = carat, y = price)) +
    geom_point() + theme_bw() +
    labs(x="Weight (carats)", y="Diamond price ($)") + # used for renaming axes labels
    ggtitle("Exploratory analysis of Diamond prices") + # used for providing title to chart
    coord_cartesian(xlim = c(0, 3), ylim=c(0, 10000)) + # used for limiting the x and y axes values boundaries
    theme(plot.title = element_text(size = 10)) #blah blah
```

```r
# histogram
diamonds %>% ggplot(aes(x = price)) +
    geom_histogram(binwidth = 500) +
    labs(x = "Price Range", y = "Total Count")
```

```r
# install.packages('gapminder')
library(gapminder)
```

```r
gapminder  %>% head
```

```r
# Timeseries Graph
gapminder %>% filter(country == 'Japan') %>% ggplot(aes(x = year, y = lifeExp)) + geom_line()
```

```r
gapminder %>% filter(country %in% c('Japan', 'India')) %>% ggplot(aes(x = year, y = lifeExp)) + geom_line(aes(color = country))
```

```r
gapminder %>% filter(country %in% c('Japan', 'India')) %>% ggplot(aes(x = year, y = lifeExp)) + 
    geom_line(aes(color = country))  +
    scale_x_continuous(breaks = seq(1950, 2021, 15)) + #setting intervals ticks for x axis
    ggtitle("Comparision of Life-Expectancy for India/Japan") +
    theme_bw()
```
